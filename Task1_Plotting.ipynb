```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = df = pd.read_excel("Employess_Cleaned1.xlsx")
```

```python
gender_counts = df['Gender'].value_counts()
total_employees = len(df)
percentages = (gender_counts / total_employees) * 100
plt.pie(gender_counts, 
        labels=[f'{label} ({perc:.1f}%)' for label, perc in zip(gender_counts.index, percentages)], 
        colors=['#ff9999','#66b3ff'],
        explode=(0.05, 0))

plt.title('Employee Gender Distribution', fontweight='bold')
plt.axis('equal') 
plt.legend(title="Gender")
plt.show()
```

```python
gender_dept_counts = df.groupby(['Department', 'Gender']).size().unstack(fill_value=0)
gender_dept_counts.plot(kind='bar', color=['#ff9999', '#66b3ff'])
plt.title('Employee Gender Distribution by Department', fontweight='bold')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.legend(title='Gender')
plt.show()
```

```python
df['Hire Date'] = pd.to_datetime(df['Hire Date'], errors='coerce')
today = pd.Timestamp.today()
df['Service Years'] = (today - df['Hire Date']).dt.days / 365.25
plt.scatter(df['Service Years'], df['Bonus %']*100)
plt.xlabel("Service Years")
plt.ylabel("Bonus Percentage (%)")
plt.title("Service Years vs. Bonus %")
plt.show()
```

```python
avg_salary_country = df.groupby('Country')['Annual Salary'].mean()
plt.bar(avg_salary_country.index, avg_salary_country.values)
plt.xlabel("Country")
plt.ylabel("Average Annual Salary ($)")
plt.title("Average Salary by Country")
plt.show()
```
